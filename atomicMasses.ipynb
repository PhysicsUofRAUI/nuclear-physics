# The Atomic Mass Evaluation Sheet
This notebook was created to make an easier interface than the text file at https://www-nds.iaea.org/amdc/ame2016/mass16.txt. I will use it throughout my Nuclear Physics course.

## Preparing the Dataframe
To create the pandas datafram I had go through the text file line by line and create a list of lists and then I was able to create a dataframe out of that.

### Large Values At the End of the file
Looking at the very large nuclei at the end of the file (and dataframe) we can see that the first number is cut off. I am assuming this is because the format of the textfile moved it over later. Eventually I will fix this, but for now looking at the 'A' value it can be seen what that number should be.

```python
import pandas as pd

titleList = []

whole_list = []

count = 0
f = open("mass16.txt", "r")
for line in f:
    if count == 37 :
        titleList.append(line[0:4].strip())
        titleList.append(line[8:9].strip())
        titleList.append(line[13:14].strip())
        titleList.append(line[17:18].strip())
        titleList.append(line[20:22].strip())
        titleList.append(line[26:27].strip())
        titleList.append(line[32:43].strip())
        titleList.append(line[50:66].strip())
        titleList.append(line[74:91].strip())
        titleList.append(line[100:111].strip())
    
        
        
    if count > 38 :
        newEntry = []
        newEntry.append(line[0:6].strip())
        newEntry.append(line[6:12].strip())
        newEntry.append(line[12:17].strip())
        newEntry.append(line[16:20].strip())
        newEntry.append(line[19:23].strip())
        newEntry.append(line[24:29].strip())
        newEntry.append(line[29:54].strip())
        newEntry.append(line[54:73].strip())
        newEntry.append(line[72:97].strip())
        newEntry.append(line[97:124].strip())
        
        whole_list.append(newEntry)
        
        
        
        
    
    count = count + 1
  
```

```python
df = pd.DataFrame.from_records(whole_list, columns=titleList)
```
